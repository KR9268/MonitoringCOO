```python
# 내용 확인(dataframe) *** 전체 ***
# ['SEC_NEGO1', 'SEC_NEGO2','SMC_NEGO1']

query = db_cursor.execute('''
                          SELECT * FROM 상공_SEC_NEGO1
                          UNION ALL
                          SELECT * FROM 상공_SEC_NEGO2
                          UNION ALL
                          SELECT * FROM 상공_SMC_NEGO1''')
cols = [column[0] for column in query.description]
pd.DataFrame.from_records(data=query.fetchall(),columns=cols).sort_values('접수일시', ascending=False)
```

```python
# 내용 확인(dataframe) ****************** 보고자료 추출용 ******************
# ['SEC_NEGO1', 'SEC_NEGO2','SMC_NEGO1']

query = db_cursor.execute('''
                    SELECT *
                        FROM (
                        SELECT * FROM 상공_SEC_NEGO1
                        WHERE 
                        접수일시 LIKE '2023-12%' 
                        AND 처리상태 LIKE '%발급완료%'
                        AND 증명서종류 = '일반(비특혜/Non-preferential) 원산지증명서'
                        UNION ALL
                        SELECT * FROM 상공_SEC_NEGO2
                        WHERE 
                        접수일시 LIKE '2023-12%' 
                        AND 처리상태 LIKE '%발급완료%'
                        AND 증명서종류 = '일반(비특혜/Non-preferential) 원산지증명서'
                        ) AS CombinedResults;
                          ''')
cols = [column[0] for column in query.description]
pd.DataFrame.from_records(data=query.fetchall(),columns=cols).sort_values('접수일시', ascending=False).to_excel('C:\\python_source\\coo_query.xlsx')
```

```python
# 파일복제 후 빈 db생성
current_date = time.strftime('%Y%m%d',time.localtime())
shutil.copyfile('Korcham_status.db', f'Korcham_status_{current_date}.db')

db_path = 'Korcham_status.db'
conn_db = sqlite3.connect(db_path)
db_cursor = conn_db.cursor()

for each in ['SEC_NEGO1', 'SEC_NEGO2','SMC_NEGO1']:
    db_cursor.execute(f"DROP TABLE IF EXISTS 상공_{each}")
    db_cursor.execute(f'''CREATE TABLE IF NOT EXISTS 상공_{each} (
                  접수번호 varchar PRIMARY KEY , 
                  증명서종류 varchar, 
                  대표Invoice varchar(10), 
                  접수일시 datetime, 
                  처리상태 varchar, 
                  Remark varchar
                  )''')
conn_db.commit()
conn_db.close()
```
